```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.optimize import minimize, differential_evolution
```

## Минимизация гладкой функции

```python
x = [i for i in np.arange(0,31,0.2)]
y = [math.sin(x / 5.0) * math.exp(x / 10.0) + 5 * math.exp(-x / 2.0) for x in x]
```

```python
plt.plot(x,y, label = 'f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)')
plt.title('График функции', fontsize = 18)
plt.ylabel('Y', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.legend(loc = 'best')
print('Здесь мы видим, что два минимума есть, ниже увидим, что мы найдем ОБА, в зависимости от начального приближения')
```

```python
def func(x):
    y = math.sin(x / 5.0) * math.exp(x / 10.0) + 5.0 * math.exp(-x / 2.0)
    return y
```

```python
for i in range(1,31,14):
    res = minimize(func, i, method='Nelder-Mead', tol=1e-6)
    print(i,'\n',res.x,'\n_________')
```

```python
i=[2,30]
for i in i:
    res = minimize(func, i, method='BFGS', tol=1e-6)
    print(i,'\n',res.x,'\n_________')
```

##  Глобальная оптимизация

```python
bounds = [(1, 30)]
result = differential_evolution(func, bounds) 
result.x, result.fun
```

```python
plt.plot(x,y, label = 'f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)')
plt.title('График функции', fontsize = 18)
plt.ylabel('Y', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.legend(loc = 'best')
print('Из двух минимумов на интервале был выбран ГЛОБАЛЬНЫЙ')
```

## Минимизация негладкой функции

```python
def func_h(x):
    y = math.sin(x / 5.0) * math.exp(x / 10.0) + 5.0 * math.exp(-x / 2.0)
    return int(y)
```

```python
x1 = [i for i in np.arange(0,31,0.2)]
y1 = [func_h(x) for x in x1]
```

```python
plt.plot(x1,y1, label = 'f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)')
plt.title('График функции', fontsize = 18)
plt.ylabel('Y', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.legend(loc = 'best')
```

```python
i=[2,3,30]
for i in i:
    res = minimize(func_h, i, method='BFGS', tol=1e-6)
    print(i,'\n',res.x,'\n_________')
```

```python
bounds = [(1, 30)]
result = differential_evolution(func_h, bounds) 
result.x, result.fun
```
