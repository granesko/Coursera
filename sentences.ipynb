```python
import numpy as np
from scipy.spatial import distance
```

```python


text = open('_3a8d746cf4d86fba2f31586f239d11fd_sentences.txt', 'r')
file = text.read()
text.close()
sent = file.splitlines()
words = file.split()
for i in  range(len(words)):
    words[i] = words[i].lower().replace('.','').replace(',','').replace(')','').replace('(','')


```

```python
words
```

```python
words1 = list(set(words))
dict_words1 = {words1[i]: i for i in range(len(words1))}
dict_words1
```

```python
a = [i.replace('.','').replace(',','').replace(')','').replace('(','').lower().split() for i in sent]
```

```python
a[0]
```

```python
for i in range(len(a)):    
    for item in a[i]:
        if item in dict_words1:
            a[i][a[i].index(item)] = dict_words1[item] 
            
```
